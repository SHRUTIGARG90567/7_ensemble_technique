```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("mnist_train.csv")
```

```python
df
```

```python
x= df.iloc[:,1:]
y= df.iloc[:,0]
```

```python
x
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y , test_size= 0.2, random_state=42)
```

```python
x_train.shape
```

```python
y_train.shape
```

```python
x_train1, x_test1, y_train1, y_test1 = train_test_split(x_train, y_train , test_size= 0.2, random_state=42)
```

```python
x_train1.shape
```

```python
y_train1.shape
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn.ensemble import ExtraTreesClassifier
```

```python
rt = RandomForestClassifier(n_estimators=10, random_state=42)
svc= SVC(gamma="scale", random_state=42)
ext  = ExtraTreesClassifier(n_estimators=10, random_state=42)
```

```python
rt.fit(x_train1, y_train1)
svc.fit(x_train1, y_train1)
ext.fit(x_train1, y_train1)
```

```python
from sklearn.ensemble import VotingClassifier
voting_clf = VotingClassifier(
    estimators = [('rt',rt),('svc',svc),('ext',ext)],
    voting = "hard"

)

```

```python
voting_clf.fit(x_train1, y_train1)
```

```python

# calculating the performance on validation set
from sklearn.metrics import accuracy_score

for clf in (rt, svc, ext, voting_clf):
    clf.fit(x_train1, y_train1)
    y_pred = clf.predict(x_test1)
    accuracy = accuracy_score(y_test1, y_pred)
    print(clf.__class__.__name__, "accuracy:", accuracy)
```

```python
# CALCULating the performance on test set 
y_pred_test = voting_clf.predict(x_test)
accuracy_test = accuracy_score(y_test, y_pred_test)
print("Ensemble accuracy on the test set:", accuracy_test)
```
